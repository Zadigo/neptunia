```python
import requests
from bs4 import BeautifulSoup
from nltk.tokenize import BlanklineTokenizer
from lxml import etree
```

```python
url = 'https://www.unilever.com/sitemap.xml'
```

```python
response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36 Edg/112.0.1722.64'})
```

```python
parser = etree.XMLParser(encoding='utf-8')
content = etree.fromstring(response.content.decode('utf-8'), parser)
```

```python
urls = [item.text for item in content.iter()]
urls = list(filter(lambda x: x is not None and x != '\n', urls))
```
